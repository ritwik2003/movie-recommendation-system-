```python
import ast
```

```python
def change(gen):
    l = []
    for i in ast.literal_eval(gen):
        l.append(i['name'])
        
    return l
        
```

```python
movies['genres'] = movies['genres'].apply(change)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(change)
```

```python
movies.iloc[0].cast
```

```python
def change2(gen):
    l = []
    count = 0
    for i in ast.literal_eval(gen):
        if count != 3:
            l.append(i['name'])
            count+=1
        else:
            break
        
    return l
```

```python
movies['cast'] = movies['cast'].apply(change2)
```

```python
movies.head()
```

```python
def get_director(gen):
    l = []
    for i in ast.literal_eval(gen):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
        
    return l
        
```

```python
movies['crew'] = movies['crew'].apply(get_director)
```

```python
movies
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies.head()
```

```python
movies['genres']= movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies['keywords']= movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']= movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']= movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = "english")
vector = cv.fit_transform(new_df['tags']).toarray()
print(vector.shape)

input_dim = vector.shape[1]  
embedding_dim = 64 
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
model = models.Sequential()

# Encoder: Reduce dimensionality to embedding_dim
model.add(layers.Dense(256, activation='relu', input_shape=(input_dim,)))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(embedding_dim, activation='relu'))  # Embedding layer

# Decoder: Reconstruct the input
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(256, activation='relu'))
model.add(layers.Dense(input_dim, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Summary of the model
model.summary()
```

```python
def find_similar_movies(movie_title, movie_embeddings, movie_index, movie_titles, top_k=5):
    # Ensure the movie is in the dataset
    if movie_title not in movie_titles:
        print(f"Movie '{movie_title}' not found in the dataset.")
        return []
    
    # Get the index of the movie from movie_titles
    movie_idx = movie_titles.index(movie_title)

    # Compute cosine similarities
    similarities = cosine_similarity([movie_embeddings[movie_idx]], movie_embeddings)[0]
    
    # Get top_k recommended movie indices (excluding the movie itself)
    recommended_indices = similarities.argsort()[-(top_k + 1):-1][::-1]
    recommended_titles = [(movie_titles[idx], similarities[idx]) for idx in recommended_indices]

    return recommended_titles

# Ensure movie_titles is defined
movie_titles = new_df['title'].tolist()

# Example of calling the function
target_movie = 'Avatar'  # Replace with the title of the movie you're interested in
movie_idx = new_df.loc[new_df['title'] == target_movie].index[0]

# Evaluate recommendations
similar_movies = find_similar_movies(target_movie, movie_embeddings, movie_idx, movie_titles, top_k=5)
print(f"Movies similar to '{target_movie}':")
for title, score in similar_movies:
    print(f"{title}: {score:.4f}")


```
